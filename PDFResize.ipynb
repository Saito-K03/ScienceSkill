```python
import pymupdf
import pprint

reader = pymupdf.open("examplex.pdf")
pprint.pprint(reader[0].get_drawings())
reader.close()
```

```python
import pymupdf
import pprint

reader = pymupdf.open("examplex.pdf")

for page in reader:
    outerRect = []
    for element in page.get_drawings():
        if element["type"] == "s":
            if round(element["width"], 5) == 0.028:
                outerRect.append(element["items"][0][1])

    rectx0 = outerRect[0].x0
    recty0 = outerRect[0].y0
    rectx1 = outerRect[0].x1
    recty1 = outerRect[0].y1
    for element in outerRect:
        if element.x0 < rectx0:
            rectx0 = element.x0
        if element.y0 < recty0:
            recty0 = element.y0
        if element.x1 > rectx1:
            rectx1 = element.x1
        if element.y1 > recty1:
            recty1 = element.y1

    rectf = pymupdf.Rect(rectx0, recty0, rectx1, recty1)
    pprint.pprint(rectf)
    page.set_mediabox(rectf)

reader.save("examplex2.pdf")
reader.close()
```

```python
import pymupdf
import pprint

reader2 = pymupdf.open("example.pdf")

for page in reader2:
    innerRect = []
    for element in page.get_drawings():
        if element["type"] == "f":
            if element["items"][0][0] == "re":
                if element["fill"] == (1.0, 1.0, 1.0):
                    innerRect.append(element["items"][0][1])
    rect = innerRect[0]
    pprint.pprint(rect)
    pprint.pprint(page.cropbox)

    page.set_mediabox(
        pymupdf.Rect(
            page.cropbox.x0 + rect.x0,
            page.mediabox.y1 - (page.cropbox.y0 + rect.y1),
            page.cropbox.x0 + rect.x1,
            page.mediabox.y1 - (page.cropbox.y0 + rect.y0)
        )
    )
    print(page.mediabox)
    print("\n")

reader2.save("example2.pdf")
reader2.close()
```
